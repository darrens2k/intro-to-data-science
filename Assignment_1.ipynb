```python
# this is the cell to plot series from groupby

# using figsize so the labels do not overlap
plt.figure(figsize=(10,4))
plt.title("Average Price per Region")
sns.barplot(y=data2['Price'].mean().index, x=data2['Price'].mean())
plt.show()
```

```python
# this is the cell to plot series from groupby

# using figsize so the labels do not overlap
plt.figure(figsize=(10,4))
plt.title("Average Units Sold per Region")
sns.barplot(y=data2['Units Sold'].mean().index, x=data2['Units Sold'].mean())
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.title("Total Units Sold per Region")
sns.barplot(y=data2['Units Sold'].count().index, x=data2['Units Sold'].count())
plt.show()
```

```python
# this is the cell to plot series from groupby

# using figsize so the labels do not overlap
plt.figure(figsize=(10,10))
plt.title("Average Price per Product")
sns.barplot(y=data3['Price'].mean().index, x=data3['Price'].mean())
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title("Average Units Sold per Product")
sns.barplot(y=data3['Units Sold'].mean().index, x=data3['Units Sold'].mean())
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title("Total Units Sold per Product")
sns.barplot(y=data3['Units Sold'].count().index, x=data3['Units Sold'].count())
plt.show()
```

```python
# now figure out how much of each product was sold in each region

# filter only region 1
region1 = data[data["Region"]=="Region_1"]
region1Filtered = region1.groupby('Product')
region1Filtered.count()
# plot
plt.figure(figsize=(8,8))
plt.title("Amount of Each Product Sold in Region 1")
sns.barplot(x=region1Filtered['Sale Date'].count(), y=region1Filtered.count().index)
plt.xlabel("Count")
plt.show()

# maybe fill the rest in with for loop
```
