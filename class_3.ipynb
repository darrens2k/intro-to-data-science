```python
actions = data.groupby(['brand', 'event_type'])['event_type'].count().unstack('event_type')
```

```python
actions.sort_values('view',ascending=False).head(20).plot(kind='bar')
```

```python
actions['conversion'] = actions['purchase']/actions['view']
```

```python
actions.sort_values('conversion',ascending=False).head(20).plot(kind='bar')
```
